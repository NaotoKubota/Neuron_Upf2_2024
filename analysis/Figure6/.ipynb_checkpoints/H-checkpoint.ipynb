```python
import pandas as pd
pd.set_option("display.max_columns", 1000)
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
import seaborn as sns
sns.set_style('ticks')
from matplotlib_venn import venn3, venn3_circles, venn3_unweighted
```

```python
# Import gene list
sc = !cat /rhome/naotok/bigdata/Upf2_paper/DEG_Up.txt
rescued = !cat /rhome/naotok/bigdata/Upf2_paper/Bulk_Upf2P53KO_rescued_genes.txt
clip = !cat /rhome/naotok/bigdata/Upf2_paper/eCLIP/Upf1_CLiP_significant_genes_3utr.txt
```

# Venn3

```python
Set2 = sns.color_palette("Set2")

# venn diagram
g = venn3_unweighted(
    
    [set(sc), set(rescued), set(clip)],
    ("Up-regulated in\nUpf2 KO scRNA-seq", "Rescued by Trp53 KO", "Enriched in UPF1 eCLIP-seq"),
    set_colors = (Set2[0], Set2[1], Set2[2]),
    alpha = 0.8
    
)

# for text in g.set_labels:
#     text.set_fontsize(20)
for text in g.subset_labels:
    text.set_fontsize(16)

plt.tight_layout()
plt.savefig("/rhome/naotok/Upf2_paper/figure/overlap_venn3.png", dpi = 1000)
```
