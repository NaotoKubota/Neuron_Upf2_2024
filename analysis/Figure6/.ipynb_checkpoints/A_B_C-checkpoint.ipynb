```python
import pandas as pd
pd.set_option("display.max_columns", 1000)
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
import seaborn as sns
sns.set_style('ticks')
import upsetplot
from upsetplot import plot
```

# Upset plot of DEGs

- Upset plots of DEGs (b-upb/dpb<=0.05, b-upt/dpt<=0.05, cnt>=50 in one of the two samples, reproduced in both replicates) in five cell cycle phases

```python
df_10x = pd.read_csv(

    "/rhome/naotok/bigdata/Upf2_paper/Liang/DEfmn_10x.csv"

)

df_10x['sample'] = "10x"


df_IDT = pd.read_csv(

    "/rhome/naotok/bigdata/Upf2_paper/Liang/DEfmn_IDT.csv"

)

df_IDT['sample'] = "IDT"


df = pd.concat(

    (df_10x, df_IDT),
    axis = 0
    
)
```

```python
df.head()
```

```python
DEG_dict = {}
DEG_Up_dict = {}
DEG_Down_dict = {}

for STAGE in ['G2/M', 'G2', 'S', 'G1/S', 'M/G1']:

    for GROUP in ["Up", "Down"]:
        
        for SAMPLE in ["10x", "IDT"]:
            
            NAME = STAGE.replace("/", "") + "_" + GROUP + "_" + SAMPLE
         
            if GROUP == "Up":
                
                DEG = list(df[
    
                    (df['stage'] == STAGE) &
                    (df['sample'] == SAMPLE) &
                    (df['b-upb'] < 0.05) &
                    (df['b-upt'] < 0.05) &
                    ((df['cnt_x'] >= 50) | (df['cnt_y'] >= 50))

                ]['gene'])
                
            else:
                
                DEG = list(df[
    
                    (df['stage'] == STAGE) &
                    (df['sample'] == SAMPLE) &
                    (df['b-dpb'] < 0.05) &
                    (df['b-dpt'] < 0.05) &
                    ((df['cnt_x'] >= 50) | (df['cnt_y'] >= 50))

                ]['gene'])
                    
            DEG_dict[NAME] = DEG
        
        DEG_INTERSECTION = set(DEG_dict[STAGE.replace("/", "") + "_" + GROUP + "_" + "10x"]) & set(DEG_dict[STAGE.replace("/", "") + "_" + GROUP + "_" + "IDT"])
            
        if GROUP == "Up":
            
            DEG_Up_dict[STAGE.replace("/", "") + "_" + GROUP] = DEG_INTERSECTION
            
        else:
            
            DEG_Down_dict[STAGE.replace("/", "") + "_" + GROUP] = DEG_INTERSECTION
```

```python
def upset(DICT: dict, COL: list, TITLE: str):

    upset_df = pd.DataFrame()

    for KEY in DICT.keys():

        DEG_list = DICT[KEY]
    
        # Diff events for each group
        Diff_df = pd.Series(

            index = DEG_list,
            data = [True]*len(DEG_list)

        )

        upset_df = pd.concat(

            (upset_df,
            Diff_df),
            axis = 1

        ).fillna(False)

    upset_df.columns = COL

    plot(

        upset_df.set_index(list(upset_df.columns)),
        subset_size = "auto",
        show_counts = "%d",
        sort_by = 'cardinality',
        sort_categories_by = None

    )

    plt.savefig("/rhome/naotok/Upf2_paper/figure/scRNA_DEG_" + TITLE + ".png", dpi = 1000)
```

```python
upset(
    DEG_Up_dict,
    ['G2/M', 'G2', 'S', 'G1/S', 'M/G1'],
    "Up"
)
```

# Heatmap of DEGs

- Heatmap of DEGs across phases

```python
DEG_gene = []

for i in DEG_Up_dict.keys():
    
    DEG_gene += DEG_Up_dict[i]

for i in DEG_Down_dict.keys():
    
    DEG_gene += DEG_Down_dict[i]

    
DEG_gene = list(set(DEG_gene))
```

```python
diff_df = df[

    df['gene'].isin(DEG_gene)
    
][["name", "stage", "sample", "cpm_x", "cpm_y",]]

diff_df.columns = ["name", "stage", "sample", "KO", "WT",]

diff_df.loc[diff_df['stage'] == "M/G1", "stage_sort"] = 1
diff_df.loc[diff_df['stage'] == "G1/S", "stage_sort"] = 2
diff_df.loc[diff_df['stage'] == "S", "stage_sort"] = 3
diff_df.loc[diff_df['stage'] == "G2", "stage_sort"] = 4
diff_df.loc[diff_df['stage'] == "G2/M", "stage_sort"] = 5

diff_df = diff_df.sort_values(["name", "stage_sort", "sample"])
```

```python
diff_cluster_df = diff_df.pivot(

        index = "name",
        columns = ["sample", "stage"],
        values = ["KO", "WT"]

    )
```

```python
group_col = diff_cluster_df.T.reset_index()['level_0']
stage_col = diff_cluster_df.T.reset_index()['stage']
group_lut = dict(zip(group_col.unique(), sns.color_palette("Set2")[0:2][::-1]))
stage_lut = dict(zip(stage_col.unique(), sns.color_palette("viridis", n_colors = 100)[::-1][0:100:20]))
group_colors = group_col.map(group_lut)
stage_colors = stage_col.map(stage_lut)
```

```python
kws = dict(cbar_kws=dict(orientation='horizontal'), figsize=(6, 6))
g = sns.clustermap(

    diff_cluster_df,
    z_score = 0,
    col_cluster = False,
    cmap = "RdBu_r",
    xticklabels = 0,
    yticklabels = 0,
    col_colors = [group_colors, stage_colors],
    vmin = -3,
    vmax = 3,
    **kws

)

# col_colors legend
group_handles = [Patch(facecolor = group_lut[name]) for name in group_lut]
stage_handles = [Patch(facecolor = stage_lut[name]) for name in stage_lut]
first_legend = plt.legend(
    group_handles,
    group_lut,
    title = "",
    bbox_to_anchor = (0.79, 1.0),
    bbox_transform = plt.gcf().transFigure,
    loc='upper right'
)

plt.gca().add_artist(first_legend)
plt.legend(
    stage_handles,
    stage_lut,
    title = "",
    bbox_to_anchor = (0.94, 1.0),
    bbox_transform = plt.gcf().transFigure,
    loc='upper right'
)

g.ax_heatmap.set_xlabel("")
g.ax_heatmap.set_ylabel("DEGs ($\it{n}$ = " + str(len(diff_cluster_df.index)) + ")")

_x0, _y0, _w, _h = g.cbar_pos
g.ax_cbar.set_position([_x0, 0.9, g.ax_row_dendrogram.get_position().width, 0.02])
g.ax_cbar.set_title('Expression\n(row $\it{Z}$-score)', fontsize = 10)
g.ax_cbar.tick_params(axis = 'x', length = 10)

plt.savefig("/rhome/naotok/Upf2_paper/figure/scRNA_DEG_heatmap.png", dpi = 1000)
```

```python
reordered_labels = diff_cluster_df.index[g.dendrogram_row.reordered_ind].tolist()
diff_cluster_focused_df = diff_cluster_df[

    diff_cluster_df.index.isin(reordered_labels[0:15])

]

kws = dict(cbar_kws=dict(orientation='horizontal'), figsize=(6, 6))
g = sns.clustermap(

    diff_cluster_focused_df,
    z_score = 0,
    col_cluster = False,
    cmap = "RdBu_r",
    xticklabels = 0,
    yticklabels = 1,
    col_colors = [group_colors, stage_colors],
    vmin = -3,
    vmax = 3,
    **kws

)

# col_colors legend
group_handles = [Patch(facecolor = group_lut[name]) for name in group_lut]
stage_handles = [Patch(facecolor = stage_lut[name]) for name in stage_lut]
first_legend = plt.legend(
    group_handles,
    group_lut,
    title = "",
    bbox_to_anchor = (0.6, 1.0),
    bbox_transform = plt.gcf().transFigure,
    loc='upper right'
)

plt.gca().add_artist(first_legend)
plt.legend(
    stage_handles,
    stage_lut,
    title = "",
    bbox_to_anchor = (0.75, 1.0),
    bbox_transform = plt.gcf().transFigure,
    loc='upper right'
)

g.ax_heatmap.set_xlabel("")
g.ax_heatmap.set_ylabel("")

_x0, _y0, _w, _h = g.cbar_pos
g.ax_cbar.set_position([_x0, 0.9, g.ax_row_dendrogram.get_position().width, 0.02])
g.ax_cbar.set_title('Expression\n(row $\it{Z}$-score)', fontsize = 10)
g.ax_cbar.tick_params(axis = 'x', length = 10)

plt.savefig("/rhome/naotok/Upf2_paper/figure/scRNA_DEG_heatmap_focused.png", dpi = 1000)
```

# GO enrichment of DEGs

```python
def metascape(metascape_path: str, output_path: str):
    
    # import metascape results
    metascape_df = pd.read_csv(

        metascape_path

    )
    
    # Summary GO
    metascape_summary_df = metascape_df[
        metascape_df['GroupID'].str.contains("Summary")
    ]

    # -Log
    metascape_summary_df['minus LogP'] = -metascape_summary_df['LogP']
    metascape_summary_df['minus LogQ'] = -metascape_summary_df['Log(q-value)']
    metascape_summary_df = metascape_summary_df.sort_values("minus LogP")

    # number of genes
    metascape_summary_df["n_genes"] = metascape_summary_df['InTerm_InList'].str.split("/", expand = True)[0].astype(int)
    
    # top 10
    metascape_summary_df = metascape_summary_df.tail(10)

    # scatter plot
    fig = plt.figure(figsize = (8, 3))
    ax = fig.add_subplot(111)

    ax = plt.scatter(

        data = metascape_summary_df,
        y = "Description",
        x = "minus LogP",
        c = metascape_summary_df['n_genes'],
        cmap = "plasma",
        edgecolor = "black",
        s = 80

    )

    cbar = plt.colorbar()
    cbar.set_label("Number of genes")

    plt.xlabel("-Log$_{10}$($\it{P}$)")
    plt.ylabel("")
    
    sns.despine()
    plt.tight_layout()

    plt.savefig(output_path, dpi = 1000)
```

```python
metascape(

    "/rhome/naotok/bigdata/Upf2_paper/metascape/metascape_result.csv",
    "/rhome/naotok/Upf2_paper/figure/scRNA_DEG_metascape.png"
    
)
```
