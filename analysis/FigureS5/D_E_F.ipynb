# Cell cycle

```python
mouse_gene_df = pd.read_csv("/rhome/naotok/bigdata/BioMart/orthologs_human2mouse.txt", sep = "\t", usecols = ["Mouse gene name", "Mouse gene stable ID"])
mouse_gene_df.columns = ["gene_name", "mgene"]
mouse_gene_df
```

```python
cellcycle_gene_liang_df = pd.read_csv(

    "/rhome/naotok/bigdata/Upf2_paper/CellCycle/cycle_mgene",
    sep = "\t",
    usecols = ["phase", "mgene"]

)

cellcycle_gene_liang_df = cellcycle_gene_liang_df.dropna()
cellcycle_gene_liang_df = pd.merge(cellcycle_gene_liang_df, mouse_gene_df, on = "mgene")
unique_genes = cellcycle_gene_liang_df.groupby("gene_name", as_index = False).count()[
    cellcycle_gene_liang_df.groupby("gene_name", as_index = False).count()["phase"] == 1
]["gene_name"]
cellcycle_gene_liang_df = cellcycle_gene_liang_df[cellcycle_gene_liang_df["gene_name"].isin(unique_genes)]
```

```python
cellcycle_gene_liang_df.groupby(["phase"]).count()
```

```python
adata_raw = adata.raw.to_adata()
adata_raw_df = adata_raw.to_df()
```

```python
phase_genes_dict = {}
for phase in cellcycle_gene_liang_df["phase"].unique():
    
    print(phase)
    
    gene_list = [x for x in cellcycle_gene_liang_df[cellcycle_gene_liang_df["phase"] == phase]["gene_name"].tolist() if x in adata_raw_df.columns]
    phase_genes_dict[phase] = gene_list
```

```python
cellcycle_gene_liang_list = cellcycle_gene_liang_df["gene_name"].tolist()
cellcycle_gene_liang_list = [x for x in cellcycle_gene_liang_list if x in adata_raw_df.columns]
```

```python
adata_raw_df = adata_raw_df[cellcycle_gene_liang_list]
```

```python
adata_raw_corr_df = adata_raw_df.corr()
```

```python
adata_raw_corr_df.shape
```

```python
phase_mean_df = pd.DataFrame()
filtered_genes_dict = {}
for phase in cellcycle_gene_liang_df["phase"].unique():
    
    print(phase)
    
    filtered_genes = []
    
    adata_raw_corr_phase_df = adata_raw_corr_df[adata_raw_corr_df.index.isin(phase_genes_dict[phase])][phase_genes_dict[phase]]
    adata_raw_corr_otherphase_df = adata_raw_corr_df[~(adata_raw_corr_df.index.isin(phase_genes_dict[phase]))][phase_genes_dict[phase]]
    
    for gene in phase_genes_dict[phase]:
        
        corr_samephase = adata_raw_corr_phase_df[gene][adata_raw_corr_phase_df[gene].index != "gene"].tolist()
        corr_otherphase = adata_raw_corr_otherphase_df[gene].tolist()
        average_samephase = sum(corr_samephase) / len(corr_samephase)
        average_otherphase = sum(corr_otherphase) / len(corr_otherphase)

        # positive average correlation with other genes in the same phase
        # larger mean correlation with other genes within the same phase than correlation with genes from different phases
        if (average_samephase > 0) and (average_samephase > average_otherphase):
            
            filtered_genes.append(gene)
            
    filtered_genes_dict[phase] = filtered_genes
    adata_filtered_df = adata_raw[:, adata_raw.var.index.isin(filtered_genes)].to_df()
    phase_mean_tmp_df = pd.DataFrame(adata_filtered_df.mean(axis = 1), columns = [phase])
    if phase_mean_df.empty:
        
        phase_mean_df = phase_mean_tmp_df
        
    else:
        
        phase_mean_df = pd.merge(phase_mean_df, phase_mean_tmp_df, left_index = True, right_index = True)
        
phase_mean_df = phase_mean_df.reset_index()
phase_mean_df = phase_mean_df.melt(id_vars = "index", var_name = "phase", value_name = "score")
```

```python
# phase with the highest score
phase_cell_df = phase_mean_df.sort_values(["index", "score", "phase"], ascending = [True, False, True]).drop_duplicates(subset = ["index"])
phase_cell_df.groupby("phase").count()
```

```python
phase_cell_dict = dict(zip(phase_cell_df["index"], phase_cell_df["phase"]))
```

```python
phase_ordered = ["M/G1", "G1/S", "S", "G2", "G2/M"]
adata.obs["phase_2"] = adata.obs.index.map(phase_cell_dict)
adata.obs["phase_2_ordered"] = pd.Categorical(values = adata.obs["phase_2"], categories = phase_ordered, ordered = True)
```

## Fraction changes in each cell type

```python
cell_cycle_number_dict = {}
cell_cycle_number_l = []

for g in ["WT", "KO"]:

    cell_cycle_number_dict[g] = {}
    
    for x in adata.obs["cell type"].unique():

        cell_cycle_number_dict[g][x] = {}

        for c in phase_ordered:
        
            # Cell number
            cell_cycle_number = adata.obs[(adata.obs["cell type"] == x) & (adata.obs["batch"] == g) & (adata.obs["phase_2_ordered"] == c)].shape[0]

            cell_cycle_number_dict[g][x][c] = cell_cycle_number
            
            cell_cycle_number_l += [[g, x, c, cell_cycle_number]]

cell_cycle_number_df = pd.DataFrame(cell_cycle_number_l, columns = ["batch", "cell type", "phase", "number"])
cell_cycle_number_df = cell_cycle_number_df.pivot(

    index = ["batch", "cell type"],
    columns = ["phase"],
    values = "number"
    
).reset_index()

cell_cycle_number_df
```

```python
cell_cycle_number_df["sum"] = cell_cycle_number_df["M/G1"] + \
cell_cycle_number_df["G1/S"] + \
cell_cycle_number_df["S"] + \
cell_cycle_number_df["G2"] + \
cell_cycle_number_df["G2/M"]

for x in phase_ordered:

    cell_cycle_number_df[x + " Percent"] = (cell_cycle_number_df[x] / cell_cycle_number_df["sum"])*100
```

```python
cell_cycle_percent_df = cell_cycle_number_df[["batch", "cell type", "M/G1 Percent", "G1/S Percent", "S Percent", "G2 Percent", "G2/M Percent"]]
cell_cycle_percent_df = cell_cycle_percent_df.melt(

    id_vars = ["batch", "cell type"],
    value_vars = ["M/G1 Percent", "G1/S Percent", "S Percent", "G2 Percent", "G2/M Percent"],
    value_name = "Ratio"

)

cell_cycle_percent_df["phase"] = cell_cycle_percent_df["phase"].str.rstrip(" Percent")
```

```python
for celltype in ["RGC", "IPC"]:

    g = sns.catplot(

        data = cell_cycle_percent_df[cell_cycle_percent_df["cell type"] == celltype],
        x = "phase",
        order = phase_ordered,
        y = "Ratio",
        hue = "batch",
        hue_order = ["WT", "KO"],
        palette = "Set2",
        kind = "bar",
        aspect = 1.3,
        height = 3

    )

    g.set_titles("{col_name}")
    g.set_axis_labels("", "% Cells")
    g._legend.set_title(None)

    plt.savefig(str(sc.settings.figdir) + "/cellfraction_phases_" + celltype + ".png", dpi = 1000)
```

## % RGC/IPC marker positive cells in S, G2, and G2M phases

```python
RGC_marker_gene_list = ["Pax6", "Top2a", "Nes"]
IPC_marker_gene_list = ["Dll3", "Insm1", "Eomes", "Neurog2", "Mgat5b", "Afap1", "Gadd45g", "Mfng", "Rwdd3"]
marker_gene_list = RGC_marker_gene_list + IPC_marker_gene_list

celltype_df = adata.obs[["cell type", "phase_2_ordered"]].reset_index()
celltype_df.columns = ["CB", "cell type", "phase"]
celltype_df.loc[celltype_df["phase"].isin(["S", "G2", "G2/M"]), "proliferating"] = True
celltype_df.loc[~(celltype_df["phase"].isin(["S", "G2", "G2/M"])), "proliferating"] = False
t = adata.raw.X.toarray()
adata_expression_df = pd.DataFrame(data=t, index=adata.obs_names, columns=adata.raw.var_names)
adata_expression_df = adata_expression_df[marker_gene_list].reset_index()
adata_expression_df = adata_expression_df.melt(id_vars = "index", var_name = "Gene", value_name = "Expression")
adata_expression_df.columns = ["CB", "Gene", "Expression"]
adata_expression_df = pd.merge(adata_expression_df, celltype_df, on = "CB")
adata_expression_df["batch"] = adata_expression_df["CB"].str.split("-", expand = True)[1]
adata_expression_df.loc[adata_expression_df["Expression"] != 0, "Positive"] = True
adata_expression_df.loc[adata_expression_df["Expression"] == 0, "Positive"] = False
```

```python
adata_positive_proliferating_df = adata_expression_df.groupby(["batch", "proliferating", "Gene", "Positive"], as_index = False).count()[["batch", "proliferating", "Gene", "Positive", "CB"]]
adata_positive_proliferating_df["Positive_ratio"] = (adata_positive_proliferating_df["CB"] / adata_positive_proliferating_df.groupby(["batch", "proliferating", "Gene"])["CB"].transform("sum"))*100
adata_positive_proliferating_df = adata_positive_proliferating_df[
    adata_positive_proliferating_df["Positive"] & adata_positive_proliferating_df["proliferating"]
]
adata_positive_proliferating_df.head()
```
