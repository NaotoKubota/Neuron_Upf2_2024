```python
import pandas as pd
pd.set_option("display.max_columns", 1000)
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('ticks')
import numpy as np
import sys
```

```python
liang_df = pd.read_csv(

    "/rhome/naotok/bigdata/Upf2_paper/Liang/DE_gene_july20_withTtest_fd2_2.5_3",
    sep = "\t"

)
```

```python
plot_df = liang_df[["name", "upf2_cortex_ko_ctr_pt", "upf2_cortex_ko_ctr_TPMfd", "upf2_cortex_ko_ctr_DEa"]]
```

```python
plot_df["upf2_cortex_ko_ctr_pt_log"] = -np.log10(plot_df["upf2_cortex_ko_ctr_pt"])
plot_df["upf2_cortex_ko_ctr_TPMfd_log"] = np.log2(plot_df["upf2_cortex_ko_ctr_TPMfd"])

# Up-regulated genes
plot_df.loc[plot_df["upf2_cortex_ko_ctr_DEa"] == 1, "color"] = "salmon"
# Down-regultated genes
plot_df.loc[plot_df["upf2_cortex_ko_ctr_DEa"] == -1, "color"] = "steelblue"

plot_df = plot_df.fillna({"color": "lightgrey"}).sort_values("color")
plot_df = plot_df.dropna()
```

```python
fig = plt.figure(figsize = (4, 4))
ax = fig.add_subplot(111)

ax = plt.scatter(

    data = plot_df,
    x = "upf2_cortex_ko_ctr_TPMfd_log",
    y = "upf2_cortex_ko_ctr_pt_log",
    c = plot_df["color"],
    alpha = 0.5,
    s = 8
    
)

plt.xlim(-11, 11)
plt.xlabel("Log2(Fold-change)")
plt.ylabel("-Log10(P)")

sns.despine()
plt.tight_layout()

plt.savefig("/rhome/naotok/Upf2_paper/figure/Emx_Upf2_KO_RNAseq_volcano.png", dpi = 800)
```
